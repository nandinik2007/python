```python
## matrix shape
A = np.array([
    [1,2,3],
    [4,5,6]
])
print(A.shape)
```

```python
## matrix addition
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print(A + B)
```

```python
## matrix multiplication
A = np.array([
    [1,2],
    [3,4]
])

B = np.array([
    [5,6],
    [7,8]
])

print(A @ B)
```

```python
## dot product
a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.dot(a,b))
```

```python
## transpose
A = np.array([
    [1,2,3],
    [4,5,6]
])

print(A.T)
```

```python
## dererminant
A = np.array([
    [1,2],
    [3,4]
])

print(np.linalg.det(A))
```

```python
## matrix inverse
A = np.array([
    [1,2],
    [3,4]
])

print(np.linalg.inv(A))
```

```python
## eigenvalues and eigenvectors
A = np.array([
    [4,2],
    [1,3]
])

values, vectors = np.linalg.eig(A)

print(values)
print(vectors)
```

```python
## statistics

    ## mean 
import numpy as np

data = [10,20,30,40]

print(np.mean(data))
```

```python
## median 
import numpy as np

data = [1,2,3,100]

print(np.median(data))
```

```python
## mode 
from statistics import mode

data = [1,2,2,3,3,3,4]

print(mode(data))
```

```python
## variance
import numpy as np

data = [10,20,30,40]

print(np.var(data))
```

```python
## standard deviation
import numpy as np

data = [10,20,30,40]

print(np.std(data))
```
