```python

##conditional statement

## if-statement
x = 10

if x > 5:
    print("x is greater than 5")

## if-else statement
x = 3

if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")

## if-else-else
x = 0

if x > 0:
    print("Positive number")
elif x == 0:
    print("Zero")
else:
    print("Negative number")

## nested-if-statement
x = 15

if x > 10:
    if x < 20:
        print("x is between 10 and 20")

```

```python
## control statement


## break
for i in range(1, 6):
    if i == 3:
        break
    print(i)

##continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)

## pass
for i in range(3):
    if i == 1:
        pass  # Placeholder for future code
    else:
        print(i)
```

```python
## looping satements

## 1. for lopp
for i in range(5):
    print(i)

## ex
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

## 2. while loop
count = 0
while count < 5:
    print(count)
    count += 1

## 3. nested loop
for i in range(3):
    for j in range(2):
        print(i, j)
```
