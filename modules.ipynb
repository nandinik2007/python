```python
## how to use module
import math

print(math.sqrt(16))
print(math.pi)
```

```python
##Importing Specific Items from a Module
from math import sqrt, pi

print(sqrt(25))
print(pi)

##Use this when you only need a few functions instead of the whole module.
```

```python
##Using an Alias Name for a Module
import math as m

print(m.sqrt(9))

##Aliases make the code shorter and easier to read.
```

```python
##Random Module

import random

print(random.randint(1, 10))  # Random number between 1 and 10
```

```python
##Datetime Module
import datetime

current = datetime.datetime.now()
print("Current date and time:", current)
```

```python
##OS Module
import os

print("Current directory:", os.getcwd())
```

```python
##External Modules

import requests
response = requests.get("https://api.github.com")
print(response.status_code)
```

```python
##example
# Using built-in module
import math
print(math.sqrt(9))

# Using user-defined module
# Fix for missing my_module by dynamically creating it
import types
my_module = types.ModuleType('my_module')
my_module.add = lambda a, b: a + b
import sys
sys.modules['my_module'] = my_module
print(my_module.add(3, 4))

# Using external module
import requests
print(requests.get("https://www.google.com").status_code)
```
