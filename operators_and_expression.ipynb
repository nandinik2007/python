operators

```python
### 1. Arithmetic Operators
a = 10; b = 3
print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.33
print(a // b) # 3
print(a % b)  # 1
print(a ** b)  # 1000
```

```python
# Comparison Operators Example
a = 10
b = 5

print("a == b:", a == b)   # Equal to
print("a != b:", a != b)   # Not equal to
print("a > b:", a > b)     # Greater than
print("a < b:", a < b)     # Less than
print("a >= b:", a >= b)   # Greater than or equal to
print("a <= b:", a <= b)   # Less than or equal to
```

```python
# 3.boolean  (Logical) Operators
print("3. Logical Operators")
p = True
q = False
print(f"  p = {p}, q = {q}")
print(f"  AND (p and q)OOO,,LLLLLL;/0000000: {p and q}") # Output: False
print(f"  OR (p or q): {p or q}")   # Output: True
print(f"  NOT (not p): {not p}")   # Output: False
print("\n")
```

```python
# 4. Identity Operators (is, is not)
print("5. Identity Operators")
l1 = [1, 2, 3]
l2 = [1, 2, 3]
l3 = l1
print(f"  l1 = {l1}, l2 = {l2}, l3 = l1")
print(f"  l1 is l2: {l1 is l2}")     # Output: False (different objects)
print(f"  l1 is l3: {l1 is l3}")     # Output: True (same object)
print(f"  l1 is not l2: {l1 is not l2}") # Output: True
print("\n")
```

```python
# 5. Membership Operators (in, not in)
my_list = [10, 20, 30, 40]
print(f"  my_list = {my_list}")
print(f" {20 in my_list}")     # Output: True
print(f"  {50 not in my_list}") # Output: True
```

```python
# 6. Bitwise Operators Example
a = 10  # (Binary: 1010)
b = 4   # (Binary: 0100)

print("a & b:", a & b)   # AND
print("a | b:", a | b)   # OR
print("a ^ b:", a ^ b)   # XOR
print("~a:", ~a)         # NOT
print("a << 1:", a << 1) # Left Shift
print("a >> 1:", a >> 1) # Right Shift
```

```python
# 7. Operator Precedence Example
x = 10 + 5 * 2     # Multiplication first, then addition
y = (10 + 5) * 2   # Parentheses first
z = 2 ** 3 * 2     # Exponentiation before multiplication

print("x =", x)
print("y =", y)
print("z =", z)
```

```python
# Arithmetic Augmented Assignment Example
a = 10
b = 3

a += b
print("After a += b:", a)   # 13

a -= b
print("After a -= b:", a)   # 10

a *= b
print("After a *= b:", a)   # 30

a /= b
print("After a /= b:", a)   # 10.0

a //= b
print("After a //= b:", a)  # 3.0

a %= b
print("After a %= b:", a)   # 0.0

a = 2
a **= b
print("After a **= b:", a)  # 8
```

```python
# Bitwise Augmented Assignment Example
a = 10   # (Binary: 1010)
b = 4    # (Binary: 0100)

a &= b
print("After a &= b:", a)   # 0 (0000)

a = 10  # Reset to 10
a |= b
print("After a |= b:", a)   # 14 (1110)

a = 10
a ^= b
print("After a ^= b:", a)   # 14 (1110)

a = 10
a <<= 1
print("After a <<= 1:", a)  # 20 (10100)

a = 10
a >>= 1
print("After a >>= 1:", a)  # 5 (0101)
```

## concatenation and repetition operation

```python
## concatenation
## + operators
## Strings
str1 = "Hello"
str2 = "World"
result = str1 + " " + str2
print("String Concatenation:", result)

#list
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2
print("List Concatenation:", combined)

#Tuples
t1 = (10, 20)
t2 = (30, 40)
t3 = t1 + t2
print("Tuple Concatenation:", t3)

## repetition (* operators)
##string
word = "Hi "
repeated = word * 3
print("String Repetition:", repeated)

##list
nums = [1, 2]
repeated_list = nums * 3
print("List Repetition:", repeated_list)

##tuple
t = (5, 10)
repeated_tuple = t * 2
print("Tuple Repetition:", repeated_tuple)

```

```python
##Concatenation Assignment (+=)
## Strings
text = "Hello"
text += " World"
print("After Concatenation:", text)


#list
numbers = [1, 2, 3]
numbers += [4, 5, 6]
print("After Concatenation:", numbers)


#Tuples
t = (10, 20)
t += (30, 40)
print("After Concatenation:", t)


## Repetition Assignment (*=)
##string
word = "Hi "
word *= 3
print("After Repetition:", word)


##list
nums = [1, 2]
nums *= 2
print("After Repetition:", nums)


##tuple
t = (5, 10)
t *= 3
print("After Repetition:", t)
```
