```python
## line plot
import matplotlib.pyplot as plt
days = [1, 2, 3, 4, 5]
temperature = [30, 32, 31, 35, 36]

plt.plot(days, temperature)

plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather Report")

plt.show()
```

```python
## vertical bar plot
import matplotlib.pyplot as plt

students = ["A", "B", "C"]
marks = [80, 90, 70]

plt.bar(students, marks)

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks")

plt.show()
```

```python
## horizontal bar plot
import matplotlib.pyplot as plt

students = ["A", "B", "C"]
marks = [80, 90, 70]

plt.barh(students, marks)

plt.show()
```

```python
## bar plot
import matplotlib.pyplot as plt
students = ["A", "B", "C"]
marks = [80, 90, 70]
plt.bar(students, marks)
plt.show()
```

```python
## pie chart
import matplotlib.pyplot as plt

brands = ["Samsung", "Apple", "OnePlus"]
share = [40, 35, 25]

plt.pie(share, labels=brands)

plt.title("Mobile Market Share")

plt.show()
```

```python
## Showing Percentages
import matplotlib.pyplot as plt

brands = ["Samsung", "Apple", "OnePlus"]
share = [40, 35, 25]

plt.pie(share, labels=brands, autopct='%1.1f%%')

plt.show()
```

```python
## histogram
import matplotlib.pyplot as plt

marks = [35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 85]

plt.hist(marks)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
## scatter plot
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5]
marks = [30, 40, 50, 70, 90]

plt.scatter(study_hours, marks)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")

plt.show()
```

```python
## 1. Line Plot
    ##Used to show trends.
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

sns.lineplot(x=x, y=y)

plt.show()
```

```python
## 2. Bar Plot
  ##Used to compare categories.
students = ["A", "B", "C"]
marks = [80, 90, 70]

sns.barplot(x=students, y=marks)

plt.show()
```

```python
## scatter plot
   ##Used to show relationships between two variables.
hours = [1, 2, 3, 4, 5]
marks = [30, 40, 50, 70, 90]

sns.scatterplot(x=hours, y=marks)

plt.show()
```

```python
## horizontal bar plot
  ##Used to compare categories with horizontal bars.
marks = [35, 40, 50, 60, 70, 80, 90]

sns.histplot(marks)

plt.show()
```

```python
##5. Heatmap
  ##One of the most famous Seaborn charts.
  ##Used to show relationships in matrix form

import seaborn as sns
import matplotlib.pyplot as plt

data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

sns.heatmap(data)

plt.show()
```

```python
## pair plot

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")

sns.pairplot(data)

plt.show()
```

```python
## heatmap 
import seaborn as sns
import matplotlib.pyplot as plt

data = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

sns.heatmap(data)

plt.show()
```
