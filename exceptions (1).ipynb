```python
## exceptions

## 1: Handling ZeroDivisionError
try:
    a = 10 / 0
except ZeroDivisionError:
    print("You cannot divide by zero!")
```

```python
## 2: Handling Multiple Exceptions
try:
    x = int("abc")
except ValueError:
    print("Value Error: Cannot convert string to integer!")
except TypeError:
    print("Type Error occurred!")
```

```python
## 3: Using else and finally
##else → runs if no exception occurs
##finally → runs no matter what, useful for cleanup
# Example: Using else and finally in Exception Handling

try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("Cannot divide by zero!")
except ValueError:
    print("Invalid input! Please enter a number.")
else:
    print("Division successful!")
finally:
    print("Program ended.")
```

```python
## 4: Raising Your Own Exception
##You can manually raise an exception using the raise keyword.
# Example: Raising Your Own Exception in Python

try:
    age = int(input("Enter your age: "))
    if age < 0:
        raise ValueError("Age cannot be negative!")  # manually raising an exception
    else:
        print(f"Your age is {age}. Valid input!")
except ValueError as e:
    print("Error:", e)
finally:
    print("Program ended.")
```

```python
##5: Custom Exception Class
##You can create your own exception type.
class InvalidAgeError(Exception):
    pass

age = 15
try:
    if age < 18:
        raise InvalidAgeError("You must be 18 or older!")
except InvalidAgeError as e:
    print(e)
```

```python
##Handling multiple exceptions
##You can also catch other possible errors:
try:
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))
    result = a / b
    print("Result:", result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Please enter numeric values only.")
```

type error

```python
##1: Adding incompatible types
# Example 1: TypeError
num = 10
text = "Python"

result = num + text   # ❌ Can't add int and str
print(result)
```

```python
##2: Calling a non-function
# Example 2: TypeError by calling a non-function
x = 5
x()   # ❌ Trying to call an integer like a function
```

```python
##3: Wrong argument type in a function
def add_numbers(a, b):
    return a + b

# Passing incorrect types
print(add_numbers(5, "10"))   # ❌ int + str
```

```python
##4: Handling TypeError using try-except
try:
    a = 5
    b = "hello"
    print(a + b)
except TypeError:
    print("❌ Type Error: Incompatible data types used together!")
```

```python
##5: With else and finally
try:
    a = 10
    b = "5"
    result = a + b
except TypeError:
    print("❌ Cannot add integer and string!")
else:
    print("✅ Operation successful:", result)
finally:
    print("🔚 Program finished.")
```

index error

```python
##1: Basic IndexError
# Example 1: IndexError example
fruits = ["apple", "banana", "mango"]
print(fruits[3])   # ❌ There is no index 3 (only 0, 1, 2)
```

```python
##2: Negative Index Works (within range)
fruits = ["apple", "banana", "mango"]
print(fruits[-1])   # ✅ Accesses last item: "mango"
```

```python
##3: Handling IndexError using try-except
fruits = ["apple", "banana", "mango"]

try:
    print(fruits[5])
except IndexError:
    print("❌ IndexError: Index out of range!")
```

```python
##4: Safe Access with Condition
fruits = ["apple", "banana", "mango"]
index = 2

if index < len(fruits):
    print("Fruit:", fruits[index])
else:
    print("❌ Invalid index!")
```

```python
##5: With else and finally
fruits = ["apple", "banana", "mango"]

try:
    index = int(input("Enter an index: "))
    print("Fruit:", fruits[index])
except IndexError:
    print("❌ Error: Index out of range!")
except ValueError:
    print("❌ Error: Please enter a valid number!")
else:
    print("✅ Access successful!")
finally:
    print("🔚 Program ended.")
```

key error

```python
##1: Basic KeyError

student = {"name": "Nandini", "age": 21}

print(student["grade"])   # ❌ 'grade' key does not exist

```

```python
##2: Preventing KeyError using .get()
##The .get() method returns None (or a default value) instead of raising an error.
student = {"name": "Nandini", "age": 21}

print(student.get("grade"))         # ✅ No error, returns None
print(student.get("grade", "N/A"))  # ✅ Returns default value 'N/A'
```

```python
##3: Handling KeyError using try-except
student = {"name": "Nandini", "age": 21}

try:
    print(student["grade"])
except KeyError:
    print("❌ KeyError: 'grade' key not found in dictionary!")
```

```python
##4: Using else and finally
student = {"name": "Nandini", "age": 21}

try:
    key = input("Enter key to access: ")
    print("Value:", student[key])
except KeyError:
    print("❌ Error: That key doesn’t exist!")
else:
    print("✅ Key found successfully!")
finally:
    print("🔚 Program ended.")
```

```python
##5: Checking for a key before access
student = {"name": "Nandini", "age": 21}

key = "grade"

if key in student:
    print("Value:", student[key])
else:
    print("❌ Key not found in dictionary.")

```
