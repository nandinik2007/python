```python
##1: Simple Function (No Parameters, No Return)
def greet():
    print("Hello! Welcome to Python functions.")

# Calling the function
greet()
```

```python
##2: Function with Parameters
def greet_user(name):
    print("Hello,", name, "welcome to Python!")

greet_user("Nandini")
```

```python
##3: Function with Return Value
def add_numbers(a, b):
    return a + b

result = add_numbers(10, 20)
print("Sum:", result)
```

```python
##4: Function with Default Parameter
def greet(name="Guest"):
    print("Hello,", name)

greet("Nandini")
greet()  # uses default value
```

```python
##5: Function with Multiple Returns
def calculate(a, b):
    sum_ = a + b
    diff = a - b
    return sum_, diff

result1, result2 = calculate(10, 5)
print("Sum:", result1)
print("Difference:", result2)
```

```python
##6: Function Calling Another Function
def square(n):
    return n * n

def cube(n):
    return n * square(n)

print("Square:", square(4))
print("Cube:", cube(4))
```

```python
##7: Anonymous Function (Lambda)
##A lambda function is a short, one-line function with no name.
add = lambda x, y: x + y
print("Sum:", add(5, 3))
```

```python
##8: Function with *args and **kwargs
##These allow variable-length arguments.
def show_details(*args, **kwargs):
    print("Arguments:", args)
    print("Keyword Arguments:", kwargs)

show_details(1, 2, 3, name="Nandini", age=21)
```

```python
##Real-World Example
def calculate_bill(amount, tax_rate):
    total = amount + (amount * tax_rate / 100)
    return total

price = float(input("Enter item price: "))
tax = float(input("Enter tax percentage: "))

print("Total bill amount:", calculate_bill(price, tax))
```
